## Grabbing Sample Data (*My* Top Artists)

```python
import pandas as pd
import spotipy
from dotenv import load_dotenv
import os
from spotipy.oauth2 import SpotifyClientCredentials, SpotifyOAuth
```

```python
# load environmental variables - API keys
load_dotenv()
```

```python
scope = ["user-library-read", "user-read-email", "user-top-read"]
SPOTIFY = spotipy.Spotify(auth_manager=SpotifyOAuth(scope=scope))
```

```python
results = SPOTIFY.current_user_top_artists(limit=50, time_range="short_term")
results
```

```python
top_artists = []

for item in results['items']:
    
    artist_info = {
        "name": item["name"], 
        "uri": item["uri"],
        "genres": item["genres"],
        "popularity": item["popularity"],
        "followers": item["followers"]["total"],
        "image_url": item["images"][1]["url"]
    }

    top_artists.append(artist_info)

TOP_ARTISTS = pd.DataFrame(top_artists)
TOP_ARTISTS
```

```python
TOP_ARTISTS.to_csv("/home/jasmine/PROJECTS/soniverse/sample_data/top_artists.csv", index=False)
```

```python
# trying to get aggregate audio features
## testing with bad omens first
omens_uri = TOP_ARTISTS[TOP_ARTISTS["name"] == "Bad Omens"]["uri"].values[0]
omens_uri
```

```python
omens_albums = SPOTIFY.artist_albums(omens_uri)["items"]
omens_albums
```

```python
omens_albums = [album for album in omens_albums if album["album_group"] == "album"]
omens_albums
```

```python
omens_albums_info = []

for album in omens_albums:
    album_info = {
        "name": album["name"],
        "uri": album["uri"],
        "release_date": album["release_date"],
        "spotify_url": album["external_urls"]["spotify"]
    }
    omens_albums_info.append(album_info)

omens_albums_info
```

```python
pd.DataFrame(omens_albums_info).to_csv("/home/jasmine/PROJECTS/soniverse/sample_data/bad_omens_albums.csv", index=False)
```

```python
all_track_info = []

for album in omens_albums_info: 

    album_result = SPOTIFY.album_tracks(album["uri"])["items"]
    print(f"Retrieved info for album {album['name']}")

    for track in album_result:

        track_info = {
            "name": track["name"],
            "album": album["name"],
            "uri": track["uri"],
            "track_number": track["track_number"],
            "duration_ms": track["duration_ms"],
            "explicit": track["explicit"],
            "spotify_url": track["external_urls"]["spotify"],
            "preview_url": track["preview_url"]
        }
        all_track_info.append(track_info)

all_track_info
```

```python
all_track_info = pd.DataFrame(all_track_info)
```
